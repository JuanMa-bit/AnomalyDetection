Verificaremos que todas las cedulas que compraron están en la base de datos de cedulas habilitadas

```python
s1=set(map(int,list(dfCompras['Rut en boleta'].value_counts().index)[1:]))
s2=set(map(int,list(dfCedulas['NUMERO IDENTIFICACION'].value_counts().index)))
s3=set(map(int,list(dfBoletaEmpleado['Rut en boleta'].value_counts().index)))
#Cedulas que compraron pero no estaban "habilitadas"
c1=len(s1-s2)
#Cedulas que estaban habilitadas pero no compraron
c2=len(s2-s1)
#Cedulas que estaban habilitadas y compraron
c3=len(s2&s1)
#La intersección son aquellas que tienen su casilla en 'si' en la columna 'Boleta Empleado'
print(len(s3&s2))

print('Total cedulas habilitadas para comprar: ',len(s2))
print('Total cedulas que realizaron compras: ',len(s1))
print('Cedulas que compraron pero no estaban "habilitadas: ',c1)
print('Cedulas que estaban habilitadas pero no compraron: ',c2)
print('Cedulas que estaban habilitadas y compraron: ',c3)
```

```python
dfCantidadFacturasEmpleado=pd.DataFrame(dfBoletaEmpleado.groupby('Rut en boleta')['Número de boleta SII'].nunique())
dfCantidadFacturasEmpleado.reset_index(inplace=True)
dfCantidadFacturasEmpleado=dfCantidadFacturasEmpleado.rename(columns={'Número de boleta SII':'Cantidad boletas'})
dfCantidadFacturasEmpleado
```

```python
dfCantidadTiendasEmpleado=pd.DataFrame(dfBoletaEmpleado.groupby('Rut en boleta')['Tienda'].nunique())
dfCantidadTiendasEmpleado.reset_index(inplace=True)
dfCantidadTiendasEmpleado=dfCantidadTiendasEmpleado.rename(columns={'Tienda':'CantidadTiendas'})
dfCantidadTiendasEmpleado
```

```python
dfCantidadUnidadesEmpleado=pd.DataFrame(dfBoletaEmpleado.groupby('Rut en boleta')['Ventas en Unidades'].sum())
dfCantidadUnidadesEmpleado.reset_index(inplace=True)
dfCantidadUnidadesEmpleado=dfCantidadUnidadesEmpleado.rename(columns={'Ventas en Unidades':'TotalUnidades'})
dfCantidadUnidadesEmpleado
```

```python
dfCantidadCompraEmpleado=pd.DataFrame(dfBoletaEmpleado.groupby('Rut en boleta')['Venta en $ sin Impuestos'].sum())
dfCantidadCompraEmpleado.reset_index(inplace=True)
dfCantidadCompraEmpleado=dfCantidadCompraEmpleado.rename(columns={'Venta en $ sin Impuestos':'TotalCompraPesos'})
dfCantidadCompraEmpleado
```

```python
dfCantidadPromoEmpleado=dfBoletaEmpleado.groupby('Rut en boleta').AplicaCodPromo.value_counts().unstack(fill_value=0)
dfCantidadPromoEmpleado.reset_index(inplace=True)
dfCantidadPromoEmpleado
```

```python
dfCantidadPromoTotalEmpleado=pd.DataFrame(dfBoletaEmpleado.groupby('Rut en boleta')['AplicaCodPromo'].count())
dfCantidadPromoTotalEmpleado.reset_index(inplace=True)
dfCantidadPromoTotalEmpleado=dfCantidadPromoTotalEmpleado.rename(columns={'AplicaCodPromo':'TotalPromos'})
dfCantidadPromoTotalEmpleado
```

```python
dfBoletaEmpleado1=dfBoletaEmpleado.merge(dfCantidadFacturasEmpleado,how='left',on='Rut en boleta')
dfBoletaEmpleado2=dfBoletaEmpleado1.merge(dfCantidadTiendasEmpleado,how='left',on='Rut en boleta')
dfBoletaEmpleado3=dfBoletaEmpleado2.merge(dfCantidadUnidadesEmpleado,how='left',on='Rut en boleta')
dfBoletaEmpleado4=dfBoletaEmpleado3.merge(dfCantidadCompraEmpleado,how='left',on='Rut en boleta')
dfBoletaEmpleado5=dfBoletaEmpleado4.merge(dfCantidadPromoEmpleado,how='left',on='Rut en boleta')
dfBoletaEmpleado6=dfBoletaEmpleado5.merge(dfCantidadPromoTotalEmpleado,how='left',on='Rut en boleta')
dfBoletaEmpleado6
```

Que tipo de promociones se usaron más en general

```python
import plotly.graph_objects as go
promoType=['Promo del día', 'Promo empleados']
dfFilterGeneral=dfBoletaEmpleado6[['Rut en boleta','PromoDay','Si','Tienda','TotalUnidades']].drop_duplicates()
promDia=dfFilterGeneral['PromoDay'].sum()
promEmp=dfFilterGeneral['Si'].sum()

fig = go.Figure([go.Bar(x=promoType, y=[promDia, promEmp])])

fig.show()
```

```python
TiendasEmpleados=list(dfBoletaEmpleado6['Tienda'].value_counts().index)
```

```python
import plotly.graph_objects as go


fig = go.Figure()
fig.add_trace(go.Bar(
    x=TiendasEmpleados,
    y=[dfFilterGeneral[dfFilterGeneral['Tienda']==tienda]['PromoDay'].sum() for tienda in TiendasEmpleados],
    name='Promoción del día',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=TiendasEmpleados,
    y=[dfFilterGeneral[dfFilterGeneral['Tienda']==tienda]['Si'].sum() for tienda in TiendasEmpleados],
    name='Promoción empleados',
    marker_color='lightsalmon'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
    
```

```python
import plotly.graph_objects as go


for tienda in TiendasEmpleados:
    df = dfFilterGeneral[dfFilterGeneral['Tienda']==tienda]
    
    fig = go.Figure()

    # Add traces
    fig.add_trace(go.Scatter(x=df['Rut en boleta'].to_list(), y=df['PromoDay'].to_list(),
                    mode='lines+markers',
                    name='Promoción del día'))
    fig.add_trace(go.Scatter(x=df['Rut en boleta'].to_list(), y=df['Si'].to_list(),
                    mode='lines+markers',
                    name='Promoción empleados'))
    fig.update_layout(title=tienda,
                  yaxis_zeroline=False, xaxis_zeroline=False)
    fig.show()
```
